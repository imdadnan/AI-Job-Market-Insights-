Analysis and Visualization

```python
# to check title value count
df.Job_Title.value_counts()
```

```python
# Prepare data
value_counts = df['Job_Title'].value_counts().reset_index()
value_counts.columns = ['Job_Title', 'Count']

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=value_counts,
    x='Job_Title',
    y='Count',
    palette='viridis'
)

# Customize
plt.title('Count of Each Job Title')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# to check industry value count
df.Industry.value_counts()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(
    data=df,
    y='Industry',          # horizontal bars
    order=df['Industry'].value_counts().index,
    palette='plasma'
)

plt.title('Count of Each Industry', fontsize=14, pad=20)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Industry', fontsize=12)

plt.tight_layout()
plt.show()

value_counts = df['Industry'].value_counts().reset_index()
value_counts.columns = ['Industry', 'Count']
```

```python
# to check company size value count
df.Company_Size.value_counts()
```

```python
# Prepare data
value_counts = df['Company_Size'].value_counts().reset_index()
value_counts.columns = ['Company_Size', 'Count']

# Custom colors (matching your Plotly version)
custom_colors = {'Small': 'yellow', 'Medium': 'blue', 'Large': 'green'}

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    value_counts['Count'],
    labels=value_counts['Company_Size'],
    autopct='%1.1f%%',  # Show percentages
    startangle=90,      # Rotate starting position
    colors=[custom_colors[size] for size in value_counts['Company_Size']]  # Apply custom colors
  )

# Add title and style
plt.title('Distribution of Company Sizes', pad=18)
plt.tight_layout()
plt.show()
```

```python

# top job locations
top_loc = df['Location'].value_counts().head(20)

plt.figure(figsize=(10,6))
top_loc.plot(kind='bar')
plt.title("Top Job Locations")
plt.xlabel("Location")
plt.ylabel("Number of Job Postings")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#salary distribute by location
salary_col = "Salary_USD"

location_salary = df.groupby("Location")[salary_col].median().sort_values(ascending=False).head(15)

plt.figure(figsize=(10,6))
sns.barplot(
    x=country_salary.values,
    y=country_salary.index,
    palette="viridis"
)
plt.title("Median Salary by Location")
plt.xlabel("Median Salary (USD)")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

```python
# AI adoption distributed by location
location_counts = df.groupby("Location")["AI_Adoption_Level"].value_counts().unstack(fill_value=0)

plt.figure(figsize=(14,7))
sns.heatmap(
    country_counts,
    cmap="coolwarm",
    linewidths=0.3,
    annot=True,
    fmt="d"
)
plt.title("Location-wise AI Adoption Level (Heatmap)")
plt.tight_layout()
plt.show()
```

```python
# required automation skills
value_counts = df['Required_Skills'].value_counts().reset_index()
value_counts.columns = ['Required_Skills', 'Count']

plt.figure(figsize=(8, 6))

colors = sns.color_palette("plasma", len(value_counts))
wedges, texts, autotexts = plt.pie(
    value_counts['Count'],
    labels=value_counts['Required_Skills'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    pctdistance=0.85,
    textprops={'fontsize': 10}
)

plt.axis('equal')
plt.title('Distribution of Required Skills', pad=20, fontsize=14, loc='center')
plt.tight_layout()

plt.legend(
    wedges,
    value_counts['Required_Skills'],
    title="Skills",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.show()
```

```python
# to create violin plot of salary
plt.figure(figsize=(10, 6))

sns.violinplot(
    x=df["Salary_USD"],
    inner="quartile"
)

plt.title("Salary (USD) â€“ Violin Plot", fontsize=16)
plt.xlabel("Salary (USD)")

plt.tight_layout()
plt.show()
```

```python
# salary distribution
plt.figure(figsize=(10, 6))

sns.histplot(
    df["Salary_USD"],
    kde=True,
    bins=40,
)

plt.title("Salary (USD) Distribution", fontsize=16)
plt.xlabel("Salary (USD)")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# to create pie chart for remote friendly job distribution
plt.figure(figsize=(8, 6))

value_counts = df["Remote_Friendly"].value_counts()

colors = sns.color_palette("magma", len(value_counts))
explode = [0.07] * len(value_counts)

plt.pie(
    value_counts,
    labels=value_counts.index,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    startangle=90,
    wedgeprops={'edgecolor':'white', 'linewidth':1}
)

plt.title("Remote_Friendly Jobs", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# to create location wise remote friendly jobs
pivot = df.pivot_table(
    index="Location",
    columns="Remote_Friendly",
    aggfunc="size",
    fill_value=0
)

plt.figure(figsize=(14, 7))
sns.heatmap(
    pivot,
    cmap="crest",
    annot=True,
    fmt="d",
    linewidths=0.4
)

plt.title("Remote_Friendly Jobs by location", fontsize=14)
plt.xlabel("Job Type")
plt.tight_layout()
plt.show()
```

```python
# to check job growth projection count value
df.Job_Growth_Projection.value_counts()
```

```python
# to create Top 10 Salaries by Location with Remote Friendliness
plt.figure(figsize=(14, 6))

sns.scatterplot(
    y=df["Salary_USD"].head(50),
    x=df["Location"].head(50),
    hue=df["Remote_Friendly"],
    palette='bright', s=100
)

plt.title('Top 10 Salaries by Location with Remote Friendliness', fontsize=16)
plt.xlabel('Location', fontsize=14)
plt.ylabel('Salary (USD)', fontsize=14)

plt.xticks(rotation=45, ha='right')

plt.legend(title='Remote Friendly', bbox_to_anchor=(1, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
